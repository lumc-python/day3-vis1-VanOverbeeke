```python
%pylab inline
import os
import pandas as pd
import numpy as np
os.chdir('C:/Users/User/Documents/Werk/PythonCourse/day3-vis1-VanOverbeeke/')
data = 'data/records.fa'
```

```python
def calc_gc_percent(seq):
    """
    Return message string reporting GC% of sequence"""
    at_count, gc_count = 0, 0
    for char in seq.upper():
        if char in ('A', 'T'):
            at_count += 1
        elif char in ('G', 'C'):
            gc_count += 1
        else:
            raise ValueError(
                "Unexpected character found: {0}. Only "
                "ACTG allowed in {1}".format(char, seq))

    try:
        return gc_count * 100.0 / (gc_count + at_count)
    except ZeroDivisionError:
        return 0.0
```

```python
# Initiate empty list of results
values = []
# read fasta lines into list
lines = read_fasta(data)
# populate results one by one
for line in lines:
    values.append(calc_gc_percent(line))
```

```python
results = {'GC%':values}
df = pd.DataFrame(results)
```
